## Some operations with dataset

### Working with index

```python
#making a copy of your dataset. Recommended especially if you are modifying the original df
original_df = df.copy()
```

```python
#retrieving the indexes of the df
df.index
```

```python
# you can also save the index as a list, for example
index_list = list(df.index)
#printing the last indexes
print(index_list[-50:-1])
```

```python
# putting a custom index, for example that 
#for example, starting from 100, you need to make sure, your index fits the lenght of rows
df.index = range(100, len(df)+100)
df
```

```python
# if you want to reset the index
df.reset_index(drop=True) #if you use drop=False the index will be a new column in your dataframe
```

### Dropping row and columns and rename them

```python
# if you want to delete some rows, let's say the first ten
df = original_df
df.drop(df.index[0:10], axis = 0)
```

```python
# if you want to delete some columns by their positions
df.drop(df.columns[0:5], axis = 1)
```

```python
#you want to delete the column "age"
df.drop(columns = "age")
```

```python
#your want to drop several columns
columns_to_drop = ["age", "country", "city"]
df.drop(columns=columns_to_drop)
```

```python
# dropping all rows with a  missing values
df = original_df
df.dropna(axis=0).reset_index(drop=True)
```

```python
# dropping duplicated rows
df.drop_duplicates
```

```python
# dropping duplicates but from an specific column
df = original_df
df.drop_duplicates(subset=['category']).reset_index(drop=True) #in this case you get only rows with unique values in "category"
```

```python
#renaming columns
df.rename(columns={"countryOfCitizenship": "citizenship"}) #yes, dictionaries!!!
```

```python
#you can create dictionary before
columns_to_rename = {"countryOfCitizenship": "citizenship", "personName": "name", "industries": "sector", "tax_revenue_country_country": "taxes" }
df.rename(columns=columns_to_rename)
```

## Indexing and slicing

```python
df = original_df
```

```python
#selecting a column
df["organization"]
```

```python
#use double square brackets to be shown with df format
df[["organization"]]
```

```python
#select different columns
df[["status", "gender", "organization"]]
```

```python
#selecting rows by index position
df.iloc[2]
```

```python
# or use double square brackets
df.iloc[[2]]
```

```python
#selecting the first 20 rows with all columns
df.iloc[:20,:]
```

```python
#selecting the first 10 rows with the first five columns
df.iloc[:10, :5]
```

```python
#select the last 10 rows with the columns 3 to 8
df.iloc[-10:,3:9]
```

```python
#selecting rows by label and index
df.loc[5:25, ["category", "source"]]
```

```python
# conditional selection, for example all billionaries below 40 years old
df.loc[df["age"]<40,]
```

```python
# selecting billionaries with finalWorth Aobove 20000 and older than 50
df.loc[(df["finalWorth"] > 20000) & (df["age"] > 50),]
```

```python
#again, but I want to see only the first five rows
df.loc[(df["finalWorth"] > 20000) & (df["age"] > 50)][0:5]
```

```python
#again, but I want to see only the first five rows and first five columns excluding column "rank"
df.loc[(df["finalWorth"] > 20000) & (df["age"] > 50), df.columns[1:5]].head(5)
```

```python
#updating values
#let's say that the first three industries the correct name as Finance
df = original_df.copy()
df.loc[:3, "industries"] = "Finance"
```

```python
## changing data formats of columns
df["finalWorth"] = df["finalWorth"].astype(str)
df.info()
```
